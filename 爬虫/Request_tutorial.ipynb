- 发送带Header的请求 只需要带Header 参数， 然后传入对应的参数， Request-header 都是字典的形式 一般都只需要User-Agent就可以了，如果这还不够 们需要 把 例如cookie 带进去
- 使用Google Chrome 打开百度 右击选择检查， 查看NetWork中的www.baidu.com de Request Headers 查找 User-Agent
- 在请求过程中带上请求头

```python
headers ={'User-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/73.0.3683.86 Safari/537.36'} 
```

```python
response = requests.get(url = 'http://www.baidu.com', headers = headers)
```

```python
response.content.decode()
```

- 传递URL参数 在请求的过程中

```python
search_parm = {'wd':'python'}
response = requests.get(url = 'http://www.baidu.com', params = search_parm, headers = headers)
```

```python
response.status_code
```
